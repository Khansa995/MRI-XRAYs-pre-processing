# Subplot

```python
x = np.arange(1, 10, 0.1)
y1 = 2*x + 5
y2 = 3*x + 10
plt.subplot(1,2,1)
plt.plot(x, y1)
plt.subplot(1,2,2)
plt.plot(x, y2)
plt.show()
```

# Scatter Plot

```python
# SCATTER PLOT
x = [10, 20, 30, 40, 50, 60, 70 ,80, 90]
a = [1, 2, 3, 4, 5, 4, 3, 2, 9]
b = [5, 6, 7, 8, 3, 2, 1, 4, 9]
plt.figure(figsize=(3, 3))
#plt.scatter(x, a)
plt.scatter(x, a, s=200)
#plt.scatter(x, b)
#plt.scatter(x, b, s=500)
plt.scatter(x, b, s=500, marker='3')
plt.legend(['a', 'b'])
plt.title('X vs a and b')
plt.xlabel('X')
plt.ylabel('a&b')
plt.show()
```

# Box Plot 1

# Box Plot 2

```python
# BOX PLOT
one = [1,2,3,4,5,7,8,9]
two=[1,2,3,4,5,4,3,2,1]
three= [6,7,8,9,8,7,6,5,4]
data = list([one, two, three])
plt.violinplot(data, showmedians=True, showmeans=True)
plt.grid(True)
plt.title('Distribution of Data')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

# Pie Chart

```python
# PIE CHART
fruits = ['apple', 'mango', 'orange', 'grapes']
quantity = [30, 45, 12, 100]
#plt.pie(quantity, labels=fruits)
#plt.pie(quantity, labels=fruits, autopct='%0.1f%%')
#plt.pie(quantity, labels=fruits, autopct='%0.1f%%', explode=(0,0.1,0,0.1))
plt.pie(quantity, labels=fruits, autopct='%0.1f%%', explode=(0,0.1,0,0.1), colors=['r','g','b','y'])
plt.show()
```

# Dohnut Chart
